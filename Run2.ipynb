```python
import keras
import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df = pd.concat([df_train, df_test])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#欠損データの修正
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Fare'] = df['Fare'].fillna(df['Fare'].mode()[0])
#df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
#カテゴリカル・データの補填
df = pd.concat([df, pd.get_dummies(df['Embarked'])], axis=1)
df =pd.concat([df, pd.get_dummies(df['Sex'])], axis=1)
#pd.get_dummies(df['Cabin'])

df.head()
```

```python
df_X = df[['Age',  'Fare', 'Parch', 'Pclass', 'SibSp', 'C', 'Q', 'S', 'female', 'male']]
df_Y = df[['Survived']]
train_X = df_X[:len(df_X) - 418]
train_Y = df_Y[:len(df_Y) - 418]
test = df_X[-418:]
X_train, X_val, Y_train, Y_val = train_test_split(train_X, train_Y, test_size = 0.2, random_state=2)
```

```python
decision_tree = tree.DecisionTreeClassifier(max_depth=5)
decision_tree.fit(X_train, Y_train)
accuracy_score(decision_tree.predict(X_val), Y_val),accuracy_score(decision_tree.predict(X_train),Y_train)
```
